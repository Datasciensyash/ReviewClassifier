```python
import os
import plotly.graph_objects as go
import plotly.express as px

import pandas as pd
from pandas_profiling import ProfileReport
import numpy as np
```

```python
train_file = 'dataset/train.pkl'
test_file = 'dataset/test.pkl'
```

```python
train_df = pd.read_pickle(train_file)
test_df = pd.read_pickle(test_file)
```

Изучим длину комментариев. 

```python
train_df['Type'] = 'Train'
test_df['Type'] = 'Test'

dataset_df = pd.concat([train_df, test_df])
dataset_df['Length'] = dataset_df.loc[:, 'Text'].str.len()
```

```python
fig = px.histogram(dataset_df, x='Length', color='Type')
fig.show()

#Очевидно, что распределения практически совпадают.
```
